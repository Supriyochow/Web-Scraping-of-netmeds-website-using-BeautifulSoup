###  1. Requesting data from the website with request module

```python
#importing required modules for retrieving
import requests
import io
import urllib   #for handling urls
import urllib.request   
import urllib.parse

```

```python
#retrieving data using requests module
req = requests.get("https://www.netmeds.com/prescriptions")
```

```python
#printing the requests
res = req.text
print(res)
```

```python
print(req.status_code)
```

### 2. Installing and parsing the response into lxml format with beautiful soup 4

```python
!pip install beautifulsoup4
```

```python
#importing beautiful soup
from bs4 import BeautifulSoup
```

```python
#parsing the data
data=BeautifulSoup(res,'lxml')
```

### 3. Getting the data categorically with class name alpha-drug-list 

```python
read1=data.select(".alpha-drug-list")
```

```python
print(read1)
```

### 4. Getting all the urls by travelling from "li" and then into "a" and then into "href" data of the class alpha-drug-list

```python
#We will store all the urls into empty string
a1_urls=[]
```

```python
for i in range(len(read1)):
    x = read1[i].find_all("li")   #we jump into li
    for j in range(len(x)):
        y = x[j].find_all("a")    #we jump into a
        for k in range(len(y)):
            z = y[k].get("href")  #we jump into href
            a1_urls.append(z)
            print(z)
```

```python
#we will check whether it is imported in string or not
print(a1_urls)
```

### 5. We will parse data for each and every url and store them

```python
#We will store all the urls into empty string
all_urls=[]
for i in range(len(a1_urls)):
    req2 = requests.get(a1_urls[i])
    res2 = req2.text
    #print(res)
    data2=BeautifulSoup(res2,'lxml')
    read2=data2.select(".panel-body")
    for i in range(len(read2)):
        x = read2[i].find_all("li")   #we jump into li
        for j in range(len(x)):
            y = x[j].find_all("a")    #we jump into a
            for k in range(len(y)):
                z = y[k].get("href")  #we jump into href
                all_urls.append(z)
                print(z)
    
```

```python
#print the total number of medicines
print(len(all_urls))
```

### 6. Now we will collect the entire data of all these medicines in DataFrames

#### 6.1 Importing pandas module

```python
import pandas as pd
```

#### 6.2 Extracting Data and Storing the data in a DataFrame

```python
#We will store all the urls into empty string
#all_urls=[]
med_dic={}
med_no=0;
for i in range(len(all_urls)):
    req3 = requests.get(all_urls[i])
    res3 = req3.text
    #print(res)
    data3=BeautifulSoup(res3,'lxml')
    
    #checking any 404 pages
    check=data3.find_all('h2',{'class':'title'})
    if check==[]:
        #collecting the data
        medName_check = data3.find('h1',{'class':'black-txt'})
        if medName_check is not None:
            medName = medName_check.get_text()
        else:
            medName="N/A"
        medPriceres_check=data3.find('span',{'class':'price'})
        if medPriceres_check is not None:
            medPriceres = medPriceres_check.get_text()
            medMRP=medPriceres[5:]     #slicing the string
        else:
            medPriceres="N/A"
          
        medComp_check=data3.find('div',{'class':'drug-manu'})
        if medComp_check is not None:
            medComp = medComp_check.get_text()
        else:
            medComp="N/A"
            
        medMan_check=data3.find('span',{'class':'gen_drug'})
        if medMan_check is not None:
            medMan = medMan_check.get_text()
        else:
            medMan="N/A"
        medBPrices_check=data3.find('span',{'class':'final-price'})
        if medBPrices_check is not None:
            medBPrices = medBPrices_check.get_text()
            medBestPrice=medBPrices[16:]
        else:
            medBPrices="N/A"
        
        med_no+=1;
    
        med_dic[med_no]=[medName,medComp,medMan,medMRP,medBestPrice]
        #print(med_dic)
    
        med_df=pd.DataFrame.from_dict(med_dic, orient='index',columns=['Medicine Name','Composition','Manufacturer','MRP','Best Price'])
        print("Data inserted for ",med_no)
    else:
        i=i+1
        
        
    
```
